```python
def lampu_jalan(waktu, cuaca):
    # Default
    brightness = 0        # 0 - 100 %
    intensitas_cahaya = 0 # dalam satuan lux

    # Atur berdasarkan waktu
    if waktu == "siang":
        brightness = 20
        intensitas_cahaya = 300

    elif waktu == "sore":
        brightness = 60
        intensitas_cahaya = 600

    elif waktu == "malam":
        brightness = 90
        intensitas_cahaya = 900

    # Penyesuaian berdasarkan cuaca
    if cuaca == "hujan":
        brightness += 15
        intensitas_cahaya += 150

    if cuaca == "berkabut":
        brightness += 20
        intensitas_cahaya += 200

    # Batas maksimal biar ga lebih dari 100%
    if brightness > 100:
        brightness = 100

    return f"""
Waktu         : {waktu}
Cuaca         : {cuaca}
Brightness    : {brightness}%
Intensitas    : {intensitas_cahaya} lux
-------------------------------------------
Lampu otomatis menyesuaikan tingkat cahaya.
"""

# -----------------------------------------
# Contoh Penggunaan
print(lampu_jalan("siang", "cerah"))
print(lampu_jalan("malam", "hujan"))
print(lampu_jalan("sore", "berkabut"))
```

```python
# Data mahasiswa
nama = ["a", "b", "c", "d", "e"]
praktikum_A = [97, 60, 80, 80, 90]
praktikum_B = [95, 75, 95, 55, 99]
praktikum_C = [90, 80, 85, 75, 98]

# Hitung rata-rata tiap praktikum
rata_A, rata_B, rata_C = sum(praktikum_A)/5, sum(praktikum_B)/5, sum(praktikum_C)/5
rata_kelulusan = (rata_A + rata_B + rata_C)/3

print("=== PROGRAM NILAI PRAKTIKUM ===")

while True:
    print("\nMenu:")
    print("1. Lihat rata-rata Praktikum A, B, C")
    print("2. Lihat rata-rata kelulusan")
    print("3. Cek kelulusan mahasiswa")
    print("4. Keluar")

    pilih = input("Pilih menu (1-4): ")

    # --------------------- MENU 1 ---------------------
    if pilih == "1":
        print(f"\nRata-rata Praktikum A: {rata_A:.2f}")
        print(f"Rata-rata Praktikum B: {rata_B:.2f}")
        print(f"Rata-rata Praktikum C: {rata_C:.2f}")

    # --------------------- MENU 2 ---------------------
    elif pilih == "2":
        print(f"\nRata-rata kelulusan (acuan lulus) = {rata_kelulusan:.2f}")

    # --------------------- MENU 3 ---------------------
    elif pilih == "3":
        while True:
            cari = input("\nMasukkan nama mahasiswa (a-e): ").lower()
            if cari in nama:
                i = nama.index(cari)
                nA, nB, nC = praktikum_A[i], praktikum_B[i], praktikum_C[i]
                rata_mahasiswa = (nA + nB + nC)/3
                status = "LULUS" if nA>=rata_A and nB>=rata_B and nC>=rata_C else "TIDAK LULUS"

                print("\n=== HASIL MAHASISWA ===")
                print(f"Nama       : {cari.upper()}")
                print(f"Nilai A    : {nA}")
                print(f"Nilai B    : {nB}")
                print(f"Nilai C    : {nC}")
                print(f"Rata-rata  : {rata_mahasiswa:.2f}")
                print(f"Status     : {status}")
            else:
                print("Nama tidak ditemukan!")

            if input("\nPilih mahasiswa lain? (Y/N): ").upper() != "Y":
                break

    # --------------------- MENU 4 ---------------------
    elif pilih == "4":
        print("Program selesai.")
        break

    else:
        print("Pilihan tidak valid!")
```
